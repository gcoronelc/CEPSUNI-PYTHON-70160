# Creando 2 variables

```python
x = 10
y = 20
```

# Operando las variables

```python
suma =  x + y
resta = x - y
producto = x * y
cociente = x // y
resto = x % y
```

# Mostrar resultado

```python
print("RESULTADO")
print("Suma:",suma)
print("Resta:",resta)
print(f"Producto:{producto}")
print(f"Cociente:{cociente}")
print(f"Resto:{resto}")
```
