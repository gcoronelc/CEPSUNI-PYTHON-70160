# Aplicando Apply

```python
# DataSet
notas = pd.Series([15, 18, 12, 20, 9, 14])

# Funcion
def cond_nota(nota):
    if nota >= 11:
        return 'Aprobado'
    else:
        return 'Desaprobado'

print("Funciób creada correctamente.")
```

```python
# Clasificar notas con una función
condicion = notas.apply(cond_nota)

# Reporte
print("Notas:", notas.values)
print("Condicion:", condicion.values)
```

```python
# Clasificar notas con una función lambda
clasificacion = notas.apply(
lambda x: 'Aprobado' if x >= 11 else 'Desaprobado')

# Reporte
print("Notas:", notas.values)
print("Condicion:", condicion.values)
```

```python
print("Juntar como columnas")
df1 = pd.concat([notas, condicion], axis=1)
print(df1)
```

```python
print("Creando un DataFrame")
df2 = pd.DataFrame({
'Nota': notas,
'Estado': condicion
})
print(df2)
```

```python
df = pd.read_csv('rrhh.csv')
print(df)
# Columna calculada
df['ingreso_anual'] = df['salario'] * 12
print(df)
```

```python
def cond_salario(salario):
    if salario > 4500:
        return 'Alto'
    elif salario > 3500:
        return 'Medio'
    else:
        return 'Básico'

# Aplicando la funcion
df['nivel'] = df['salario'].apply(cond_salario)
print(df)
```

```python
# Columna condicional con .apply()
df['nivel'] = df['salario'].apply(
     lambda x: 'Alto2' if x > 4500 else ('Medio2' if x > 3500 else 'Básico2')
)
print(df)
```

```python
# Paso 1
# Crear el DataFrame de ejemplo
data = {
'estudiante': ['Ana', 'Luis', 'María', 'Carlos', 'Rosa', 'Jorge'],
'nota_1': [15, 12, None, 14, None, 11 ],
'nota_2': [18, None, 16, None, None, 13 ],
'nota_3': [None, None, None, None, None, None ],
'nota_4': [14, 15, 13, 16, 12, None ],
}

df_notas = pd.DataFrame(data)
print("DataFrame original:")
print(df_notas)
print()
```

```python
# Paso 2
# Ver el porcentaje de nulos por columna
print("Porcentaje de nulos por columna:")
print((df_notas.isnull().sum() / len(df_notas) * 100).round(1).astype(str) + " %")
print()
```

```python
# Paso 3
# Aplicar dropna con thresh al 50%
# filas × 0.5 = 3 valores válidos mínimos
thresh = len(df_notas) * 0.5
df_new = df_notas.dropna(thresh=thresh, axis=1)
print(f"thresh = len(df_notas) * 0.5 = {len(df_notas)} × 0.5 = {thresh}")
print()
print("DataFrame después de eliminar columnas con más del 50% de nulos:")
print(df_new)
```

```python
# Crear DataFrame con tipos como texto
df_local = pd.DataFrame({
'precio': ['100', '250', 'error', '300'],
'fecha': ['2024-01-10', '2024-02-15', '2024-03-20', '2024-04-25'],
'ciudad': ['Lima', 'Cusco', 'Lima', 'Arequipa']
})
print("DataFrame original:")
print(df_local)
print("\nTipos originales:")
print(df_local.dtypes)
# Convertir columna de texto a numérico
df_local['precio'] = pd.to_numeric(df_local['precio'], errors='coerce')
# errors='coerce' convierte valores no convertibles a NaN
# errors='raise' lanza error (default)
# errors='ignore' deja el valor original sin convertir
# Convertir columna de texto a fecha
df_local['fecha'] = pd.to_datetime(df_local['fecha'], format='%Y-%m-%d')
# Convertir a categórico (ahorra memoria en columnas con pocos valores únicos)
df_local['ciudad'] = df_local['ciudad'].astype('category')
# Verificar los tipos resultantes
print("\nTipos finales")
print(df_local.dtypes)
```

```python
df_enero = pd.DataFrame({'producto': ['A','B'], 'ventas': [100, 200]})
df_febrero = pd.DataFrame({'producto': ['A','C'], 'ventas': [150, 300]})
# Apilar verticalmente (eje de filas)
df_total = pd.concat([df_enero, df_febrero], ignore_index=True)
# Reporte
print("Enero:\n", df_enero)
print("Febrero:\n", df_febrero)
print("Todo:\n", df_total)
```

```python
# DataFrame 1: Información de productos
df_productos = pd.DataFrame({
'id': [101, 102],
'nombre': ['Laptop', 'Mouse']
})
# DataFrame 2: Información de precios (para los mismos productos)
df_precios = pd.DataFrame({
'precio': [1200, 25],
'stock': [10, 50]
})
# Concatenar horizontalmente (eje de columnas)
df_combinado = pd.concat([df_productos, df_precios], axis=1)
# Reporte
print("Productos:\n", df_productos)
print("\nPrecios:\n", df_precios)
print("\nResultado (Columnas unidas):\n", df_combinado)
```

```python
df_empleados = pd.DataFrame({
'id': [1, 2, 3, 4],
'nombre': ['Ana', 'Luis', 'María', 'Gustavo'],
'depto_id': [10, 20, 10, 40]
})

df_deptos = pd.DataFrame({
'depto_id': [10, 20, 30],
'depto': ['Ventas', 'Sistemas', 'Compras']
})

print("Empleados:\n", df_empleados)
print("\nDepartamentos:\n", df_deptos)
# Inner join (solo registros con coincidencia en ambos)
resultado = pd.merge(df_empleados, df_deptos, on='depto_id', how='inner')
print("Resultado:\n", resultado)
```
